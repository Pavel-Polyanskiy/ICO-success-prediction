```python
import warnings
warnings.filterwarnings('ignore')
trf = spacy.load('en_core_web_lg')
#trf_2 = spacy.load('en_core_web_trf')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Research papers/Thesis /Data/processed data/wp_text_cleaned_v2.csv')
```

```python
TAGS = ['LOC','ORG', 'PERSON', 'PRODUCT', "GPE"]
```

```python
df['location_tag_count'] = None
df['organization_tag_count'] = None
df['person_tag_count'] = None
df['product_tag_count'] = None
df['geopolitical_tag_count'] = None
```

```python
df.shape
```

```python
def count_tags(i, text):
    entities = []
    labels = []
    doc = trf(text)
    for ent in doc.ents:
        entities.append(ent.text)
        labels.append(ent.label_)

    df['location_tag_count'][i] = labels.count(TAGS[0])
    df['organization_tag_count'][i] = labels.count(TAGS[1])
    df['person_tag_count'][i] = labels.count(TAGS[2])
    df['product_tag_count'][i] = labels.count(TAGS[3])
    df['geopolitical_tag_count'][i] = labels.count(TAGS[4])

    print(f"[INFO]: {i} is done!")

with concurrent.futures.ThreadPoolExecutor() as executor:
    for i, text in enumerate(df['text'].values):
        executor.submit(count_tags, i, text)
```

```python
df.text = df.text.apply(lambda text: str(text).replace(" - ", ""))
```

```python
df.to_csv('/content/drive/MyDrive/Research papers/Thesis /Data/processed data/NER_v1.csv', index = None)
```
